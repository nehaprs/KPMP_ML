```python
ID_COL = 'Participant_ID_norm' 

all_ids = pd.Series(dtype='object')

for name, df in {
    'AU5812_2022': df_AU5812_2022,
    'AU5812_2024': df_AU5812_2024,
    'MSDQ120_2022': df_MSDQ120_2022,
    'MSDQ120_2024': df_MSDQ120_2024,
    'UHPLC_2024':   df_UHPLC_2024
}.items():
    ids = df[ID_COL].astype(str).str.replace("\xa0", "").str.strip()
    all_ids = pd.concat([all_ids, ids], ignore_index=True)

unique_patient_ids = pd.DataFrame({ID_COL: all_ids.unique()})
unique_patient_ids.head()
```
